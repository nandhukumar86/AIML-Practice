# Basic Imports 

```python
import skimage.io as io
import matplotlib.pyplot as plt
import numpy as np
```

# Read Image 

```python
img = io.imread('parrot.jpeg')

plt.imshow(img)
```

```python
img.shape
```

# Plot RGB Channel 

```python
# subplot of grid m x n.
fig, axes = plt.subplots(1,3,figsize=(30,10)) # set ax=axes[m][n] in specific plots parameters

axes[0].imshow(img[:,:,0],cmap='Reds')
axes[1].imshow(img[:,:,1],cmap='Greens')
axes[2].imshow(img[:,:,2],cmap='Blues')
```

# Import OpenCV 

```python
import cv2
```

## Convert 3 Channel to 1 Channel

```python
greyimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(greyimg)
```

### Edge Detection kernel initialization

```python
edgeDetectorKernel = np.array([[-2,0,-2],[0,8,0],[-2,0,-2]])
```

#### Apply Kernal through the filters

```python
plt.imshow(cv2.filter2D(img, -1, edgeDetectorKernel))
```

```python
blurKernel = np.ones((10,10))/100
```

```python
plt.imshow(cv2.filter2D(img, None, blurKernel))
```
