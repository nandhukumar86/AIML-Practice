```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
sns.set(style="darkgrid")
```

```python
cars = pd.read_csv('mtcars.csv')
```

```python
cars.head(3)
```

```python
sns.distplot(cars['mpg'])
plt.show()
```

```python
sns.jointplot(cars['disp'], cars['wt'])
# sns.jointplot(cars['disp'], cars['wt'], kind = 'hex')
```

```python
sns.pairplot(cars[['disp','wt']])
#sns.pairplot(cars[['disp','wt','cyl']]) # and so on..
```

```python
sns.stripplot(cars['cyl'], cars['mpg'])
```

```python
sns.barplot(cars['cyl'], cars['mpg'])
```

```python
sns.countplot(cars['gear'], hue= cars['carb'])
```

```python
sns.lmplot('mpg', 'disp', cars)
#sns.lmplot('mpg', 'disp', cars, hue = 'am')
```

```python
sns.factorplot(x = 'cyl', y = 'disp', hue='gear', col= 'am', data = cars, kind = 'bar')
```
