```python
import cv2
import skimage.io as io

import os
```

```python
folder = '/content/drive/MyDrive/Cars Images'

images = []

for f in os.listdir(folder):
  img = io.imread(folder+'/'+f)
  print(f, img.shape);
  images.append(img)
```

Images are of different sizes, two are HD quality images as the pixel are huge.

```python
import pandas as pd
import numpy as np
images = np.array(images)
```

```python
# subplot of grid m x n.
fig, axes = plt.subplots(15,1,figsize=(120,60))

for i in range(15):
  axes[i].imshow(images[i])

plt.show()
```
